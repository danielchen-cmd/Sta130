```python
import numpy as np
import matplotlib.pyplot as plt

# Original data
total_people = 80
correct_guesses = 49
incorrect_guesses = total_people - correct_guesses

# Number of bootstrap samples
n_bootstraps = 10000

# Original proportion of correct guesses
observed_proportion = correct_guesses / total_people

# Create an array of 1's (correct guesses) and 0's (incorrect guesses)
data = np.array([1] * correct_guesses + [0] * incorrect_guesses)

# Initialize a list to store bootstrap proportions
bootstrap_proportions = []

# Bootstrapping
for _ in range(n_bootstraps):
    # Resample data with replacement
    bootstrap_sample = np.random.choice(data, size=total_people, replace=True)
    # Compute the sample proportion of correct guesses
    sample_proportion = np.mean(bootstrap_sample)
    bootstrap_proportions.append(sample_proportion)

# Calculate 95% confidence interval
confidence_interval = np.percentile(bootstrap_proportions, [2.5, 97.5])

# Plotting the bootstrap distribution of sample proportions
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_proportions, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2, label=f'Observed Proportion: {observed_proportion:.3f}')
plt.axvline(confidence_interval[0], color='blue', linestyle='dotted', linewidth=2, label=f'95% CI Lower Bound: {confidence_interval[0]:.3f}')
plt.axvline(confidence_interval[1], color='green', linestyle='dotted', linewidth=2, label=f'95% CI Upper Bound: {confidence_interval[1]:.3f}')
plt.title('Bootstrap Distribution of Sample Proportions (Correct Guesses)')
plt.xlabel('Sample Proportion of Correct Guesses')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Display the confidence interval
print(f'95% Confidence Interval: {confidence_interval[0]:.3f} to {confidence_interval[1]:.3f}')

```
